***Part 1 - Decoposition***

```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose
```

```python
data = pd.read_csv('AirPassengers.csv')
```

```python
data.head(2)
```

Renaming Fields

```python
data.columns = ['month', 'n_passengers']
```

```python
data.head(1)
```

```python
data.describe()
```

Set Index

```python
data = data.set_index('month')
```

```python
data.head(1)
```

```python
data.index = pd.to_datetime(data.index)
```

Plotting

```python
data.plot()
plt.show()
```

Decomposition

```python
#Aditive

addtive_decomposition = seasonal_decompose(data, model='Addtive')
```

```python
addtive_decomposition.plot()
plt.show()
```

```python
#Multiplicative

multiplicative_decomposition = seasonal_decompose(data, model='Multiplicative')
```

```python
multiplicative_decomposition.plot()
plt.show()
```
